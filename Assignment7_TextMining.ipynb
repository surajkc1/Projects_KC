```python
import nltk #importing nltk
nltk.download('all')
```

A strong Python module called NLTK offers a variety of effective natural language algorithms. It has a strong community, is open source, free, and extensively documented. The most popular algorithms, including part-of-speech tagging, stemming, sentiment analysis, topic segmentation, and named entity identification, are all included in NLTK. NLTK assists the computer with text analysis, preprocessing, and comprehension.

The act of converting unstructured text into a structured format with the purpose of identifying significant patterns and fresh insights is known as text mining, also known as text data mining.

```python
text = '''This year’s Kentucky GIS conference saw a large
turnout from Kentucky State University. The Kentucky Association of Mapping Professionals (KAMP)
hosted the 2022 Kentucky GIS conference in Lexington, Kentucky from September 27-29, 2022. On
the first day of the conference, various hands-on
workshops aimed at students and young professionals were held. The second and third days were
dominated by presentations on various elements
of GIS, mapping, and remote sensing. With such
a large number of professionals and academics in
attendance, the event served as a wonderful networking opportunity. '''
```

The dataset/text in this project is taken from a monthly college publication of KSU. The text data is unstructured and hapazhard. We want to transfor these unstructured texts into a structured format to identify meaningful patterns and new insights.

```python
lower_text = text.lower()
```

```python
word_tokens = nltk.word_tokenize(lower_text)
```

The initial stage in text analytics is tokenization. Tokenization is the process of dividing a text paragraph into smaller units, such as words or sentences. A token is a single object that serves as the foundation of a phrase or paragraph.

```python
word_tokens
```

```python
sent_token = nltk.sent_tokenize(lower_text)
```

```python
from nltk.corpus import stopwords  #importing stopwords command
stop_words=set(stopwords.words("english"))
print(stop_words)
```

Stopwords are seen as textual clutter. Stop words like "is," "am," "are," "this," "a," "an," "the," etc. may appear in text.

Creating a list of stopwords in NLTK and filtering out my list of tokens from this list.

```python
from nltk.corpus import stopwords
stopword = stopwords.words('english')

word_tokens = nltk.word_tokenize(lower_text)
removing_stopwords = [word for word in word_tokens if word not in stopword] ##removing stopwords
```

```python
removing_stopwords #removing stopwords
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer #importing lemmatizer command

wordnet_lemmatizer = WordNetLemmatizer()

lemmatized_word = [wordnet_lemmatizer.lemmatize(word) for word in removing_stopwords]
```

```python
lemmatized_word
```

When words are lemmatized, they are reduced to their basic words, which are lemmas in the proper sense. With the use of vocabulary and morphological analysis, it converts root words. Typically, lemmatization is more complex than stemming. Stemmer analyzes a single word without taking context into account. For instance, the lemma of the word "better" is "good." This needs a dictionary search, so stemming will miss it.

```python

from nltk.stem import SnowballStemmer  #importing stem commands

snowball_stemmer = SnowballStemmer('english')

stemmed_word = [snowball_stemmer.stem(word) for word in removing_stopwords]
```

```python
stemmed_word
```

Stemming is a linguistic normalizing procedure that strips words of their derived affixes or reduces them to their word roots. For instance, the words connection, connected, and linking word can be reduced to the word "connect".

```python
from nltk import FreqDist

freq = FreqDist(removing_stopwords)
```

```python
freq.most_common(5)
```

```python
freq.most_common(10)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(freq.most_common(100))
df.columns = ['word','freq']
```

```python
import wikipediaapi #importing commands
wiki_wiki = wikipediaapi.Wikipedia('en')
```

```python
page_py = wiki_wiki.page('Kentucky State University') #accessing Kentucky State University from wikipedia
```

```python
from nltk.corpus import stopwords
stopword = stopwords.words('english')


input_text = page_py.text
lower_text = input_text.lower()


word_tokens = nltk.word_tokenize(lower_text)
removing_stopwords = [word for word in word_tokens if word not in stopword]
```

```python
removing_stopwords
```

```python
from wordcloud import WordCloud
removing_stopwords_str = ' '.join(removing_stopwords)
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(removing_stopwords_str)
```

```python
from matplotlib.pyplot import figure  #importing commands
import matplotlib.pyplot as plt
figure(figsize=(6, 6), dpi=700)


plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show() #plotting data
```
