```python
# import required modules  
import pandas as pd  #importing pandas
import numpy as np   #importing numpy
import matplotlib.pyplot as plt #importing matplot
import seaborn as sns
%matplotlib inline
```

The data that we are going to handle and analyse is related to heart stroke. The data consists of several variables like age, sex, cholestrol level, rest ecg, higher bp(thalachh). The data consists of output as stroke where 0 means no stroke and 1 means patients with stroke. This data is subjected for splitting into train/test. 2-3 classification models will be used to find the best classifier and model and their performance will be evaluated.

```python
HA_data = pd.read_csv("/content/heart.csv")  #reading the data
col_names = ['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh', 'exng','oldpeak','slp','caa','thall','output']
HA_data.columns= col_names #datavisualizing
HA_data.head()
```

```python
feature_col = ['sex', 'trtbps', 'chol', 'fbs', 'thalachh', 'age', 'exng']
X= HA_data[feature_col] #features
Y= HA_data.output #Target variable
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split  #importing commands
 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split( 
                        X,Y,test_size = 0.20, random_state = 0)   ## training and testing the data set
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_train,Y_train)
```

```python
#predicting the output for out test set
y_pred=logreg.predict(X_test)
y_pred   #predicting the model
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, Y_train)
```

```python
y_pred2=gnb.predict(X_test)
y_pred2 #predicting model
```

```python
df_out = X_test.copy()

df_out['LABEL'] = Y_test
df_out['PREDICTED_LOGREG'] = y_pred
```

```python
df_out['PREDICTED_NB'] = y_pred2
```

```python
df_out
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred)) #generating report of the logistic regression model
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred2))  #generating report/performance of Gaussian model
```

```python
from sklearn.svm import SVC 
```

```python
model = SVC() 
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test) 
print(classification_report(Y_test, predictions))  #Generating report of SVC model
```

```python
ypred3= model.predict(X_test)
ypred3 #predicting the model
```

```python
df_out['PREDICTED_SVC'] = ypred3
```
