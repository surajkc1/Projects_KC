4 - Using Python and your preference of statistical package(s), create a linear regression model using any combination of columns (ignore lat/long) to predict median_house_value.

Include plots of data with your regression line
Provide interpretation of the model
Assess the performance of your regression model
Use your model to make a prediction of median_house_value

```python
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
```

```python
X = data[['median_income']].values
y = data['median_house_value']
model = LinearRegression()
clf = model.fit(X, y)
print('Intercept : ', clf.intercept_)
print('Coefficient: ', clf.coef_[0])
predictions = model.predict(X)
print(X)
```

The relationship between median income and median house value is being calculated. Here, median income is independent variable X and median house value is dependent variable y. Simple linear regression model will be built so the median house value on basis of median income can be predicted.

```python
print(f"intercept: {model.intercept_}")  ###Calculating the intercepts
print(f"slope: {model.coef_}")
```

Equation: we have calculated slope and intercept and now can construct the equation on basis of these values. Here y value can be calculated i.e. median house value using x i.e. median income as following:
y= 42054.07487406(x)+ 43980.62815629659.

```python
y_pred = model.predict(X)   ## Using the model for predictions with  existing  data
print(f"predicted response:\n{y_pred}")
```

```python
metrics.mean_squared_error(y, predictions)
```

```python
r_sq = model.score(X,y) ### Coefficient of determination means what amount of variation in Y(median_house_value) can be explained by the dependence on X(median_income)
print(f"coefficient of determination: {r_sq}")
```

```python
plt.scatter(X, y)     #plotting the regression line and scatter plots
plt.plot(X, predictions, color='red')
plt.show()
```

The Mean Squared Error(MSE) value has been found 7012439548.9144 and R2 value is 0.478684932375828. It means 47.8% of the variation in y i.e. median house value can be explained by X i.e. Median income. As the value of R2 is less than 50% and MSE value is also quite high, this relationship cannot be explained as "Strong".
